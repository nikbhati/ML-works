```python
unlocked_cols = ['Call ID','Unlocked Date','Investor Name','Investor Phone','Scrip',
                 'Advisory Fee','Status','Advisor Name','P/L','Investment','P/L','Status','Type (Intraday/Positional)']
```

```python
import pandas as pd
import numpy as np
```

```python
unlocked_data=pd.read_csv('UnlockedCalls.csv')
```

```python
unlocked_data.head()
```

```python
all_data=pd.read_csv('AllCallsData.csv')
```

```python
all_data.head()
```

```python
unlocked_data=pd.merge(all_data,unlocked_data)
```

```python
unlocked_data.head()
```

```python
unlocked_data.shape
```

```python
unlocked_callType=unlocked_data.pivot_table(index=['Investor Phone'],columns=['Call Type'],values='P/L',aggfunc=np.sum)
```

```python
unlocked_callType
```

```python
nifty_fut_positional_PnL=unlocked_callType['NIFTY-future-positional']
```

```python
nifty_fut_positional_PnL
```

```python
similar_calls=unlocked_callType.corrwith(nifty_fut_positional_PnL)
```

```python
similar_calls=similar_calls.dropna()
```

```python
df=pd.DataFrame(similar_calls)
```

```python
df=df.sort_values(by=[0],ascending=False)
```
